Exercice 2 : Fusion de listes tri√©es

```python
def fusion_liste(liste1, liste2):
    if not liste1:
        return liste2
    if not liste2:
        return liste1

    if liste1[-1] > liste2[-1]:
        return fusion_liste(liste1[:-1], liste2) + [liste1[-1]]
    else:
        return fusion_liste(liste1, liste2[:-1]) + [liste2[-1]]
```

```python
liste1 = [3, 1, 4, 7, 9, 10, 11]
liste2 = [2, 5, 6, 1,6, 13]

liste1.sort()
liste2.sort()

Fusion_listes = fusion_liste(liste1, liste2)
print(Fusion_listes)
```
